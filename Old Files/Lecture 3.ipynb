```python
##Takes a clean character dict and a VAD vocabulary values dict and returns a list of ANEW scores
def extract_ANEW(clean_pages,vocab_weights):
    from sklearn.feature_extraction.text import CountVectorizer
    import numpy as np
    vectorize = CountVectorizer(min_df=1,vocabulary=vocab_weights.keys())
    X = vectorize.transform(clean_pages.values()).toarray()
    Z=np.dot(X,vocab_weights.values())
    return np.transpose(np.transpose(Z)/np.sum(X,1))
```

```python
##Takes a clean character dict and a VAD vocabulary values dict and returns a dict of scores relating to pages
def extract_ANEW_dict(clean_pages,vocab_weights):
    values = extract_ANEW(clean_pages,vocab_weights)
    return dict(zip(clean_pages.keys(),values))
```
